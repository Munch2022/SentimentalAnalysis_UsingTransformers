```python
from transformers import pipeline
```

```python
sent_analyzer = pipeline('sentiment-analysis', model='arpanghoshal/EkmanClassifier')
```

```python
sent_analyzer("Thanks for your help")
```

```python
sent_analyzer("you are a useless bot")
```

```python
# Using Cust_reviews.csv from my system(have end user replies on bot responses)

import pandas as pd
reviews= pd.read_csv("/content/Cust_reviews.csv")
```

```python
reviews.head()
```

```python
# writing a function so that I would implement it on all the rows/reviews

def senti_analysis(text):
  emotion= sent_analyzer(text)[0]['label']
  return emotion


```

```python
senti_analysis("how are you")    # calling function with a text
```

```python
senti_analysis("i am so mad at your responses")
```

```python
reviews['Sentimental_analysis']= reviews['EndUser response'].apply(senti_analysis)
```
